# Categoria

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from create_model import model_pycaret,call_model
```

```python
dataset = pd.read_csv("../dataset/Soles/categoria.csv",index_col=0,parse_dates=["Fecha"])
```

```python
dataset.head()
```

```python
for i in dataset.Categoria.unique():
    subset = dataset[dataset['Soles']==i].drop('Categoria',axREis=1).set_index('Fecha')
    std = np.std(subset['Soles'])
    subset = subset.resample('B').sum()
    subset['moving_average'] = subset['Soles'].rolling(30).mean()
    plt.plot(subset['Soles'])
    plt.plot(subset["moving_average"] + std)
    plt.plot(subset["moving_average"] - std)
    plt.plot(subset["moving_average"] + 2*std)
    plt.plot(subset["moving_average"] - 2*std)
    plt.plot(subset["moving_average"])
    plt.title(i)
    plt.xticks(rotation=90)
    plt.show()
```

```python
model_pycaret(dataset,target="Soles",level="Categoria")
```

```python
concat_df = call_model(dataset,target="Soles",level="Categoria")
```

```python
for i in concat_df['time_series'].unique():
    sub_df_1 = concat_df[concat_df['time_series']==i][['Label']]
    sub_df_2 = dataset[dataset['Categoria'] == i].set_index('Fecha')[['Soles']]
    sub_df_2 = sub_df_2.resample('B').sum()
    sub_df = pd.concat([sub_df_1,sub_df_2],axis=1)
    plt.plot(sub_df[['Soles', 'Label']])
    plt.title(i)
    plt.xticks(rotation=90)
    plt.show()
```
